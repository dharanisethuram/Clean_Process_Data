```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('keycleanUser_Ment.csv')
# This data haa no NaN values of mentions
```

```python
# Username's are put into a list!
du = df['Username']
#du = du[0:10000]
arr = [100] #adjust to needed 
uname = []
for row in du:
    uname.append(row)
print (uname)
```

```python
dm = df['Mentions']

#print(m)
#arr[100]
x = []
for row in dm:
   x.append(row)
print(x)
```

```python
# Mentions are put into a list of lists!

ment_list=[]

for i in x:
    y = i.split(",")
    z=[]
    for j in y:
        if j!=' ':
           z.append(j)
        
        
    ment_list.append(z)

ment_list
    
```

```python
ment_list
uname
conpair=[]    
    
for j, jval in enumerate(ment_list):
    for k in jval:
        for i, ival in enumerate(uname):
            if k==ival:
                iuser=j
                jmen=i
                if uname[iuser] in ment_list[jmen]:
                    conpair.append([uname[iuser],k])
                
            

            
print(conpair)

            
```
