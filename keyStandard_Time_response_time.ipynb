```python
for index,row in df2.iterrows():
    #Getting username1 and username2
    u1=row['Username1']
    u2=row['Username2']
    #Now getting when user1 mentioned user2
    post_time1=df1.loc[(df1['Username']==u1) & (df1['Mentions'].str.contains(u2))]['Post Time - Standard']
    p1=post_time1.tolist()
    
    #Now getting when user2 mentioned user1
    post_time2=df1.loc[(df1['Username']==u2) & (df1['Mentions'].str.contains(u1))]['Post Time - Standard']
    p2=post_time2.tolist()
    df2.set_value(index,'Post_Time_user1',p1[0])
    df2.set_value(index,'Post_Time_user2',p2[0])
    
    #Finding response time between the two by checking which is greater
    
    if p1[0]>p2[0]:
        response=p1[0]-p2[0]
        df2.set_value(index,'Response_time',response)
    else:
        response=p2[0]-p1[0]
        df2.set_value(index,'Response_time',response)
    
        
        
    
    
    
    
   
   
    
```

```python
#Finding Average of the Reponse Time
import functools
import operator
def avg_datetime(series):
    dt_min = series.min()
    deltas = [x-dt_min for x in series]
    return dt_min + functools.reduce(operator.add, deltas) / len(deltas)

```

```python
list_time=df2['Response_time']
```

```python
print(avg_datetime(list_time))
```
